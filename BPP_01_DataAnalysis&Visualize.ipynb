```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("student_data.csv")

num_bangladesh_students = df[df['country'] == 'Bangladesh'].shape[0]
print(f"Number of students from Bangladesh: {num_bangladesh_students}")
```

```python
#Calculate the average attendance for students in the "Information Technology" department

avg_attendance_it = df[df['department'] == 'Information Technology']['attendance'].mean()
print(f"Average attendance for IT department: {avg_attendance_it}")
```

```python
#display students above cgpa 3.5 and study abroad

filtered = df[(df["cgpa"] > 3.5) & (df["abroad_study"] == "Yes")]
print("filtered:", filtered)
```
